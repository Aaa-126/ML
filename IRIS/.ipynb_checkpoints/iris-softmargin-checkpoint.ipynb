```python
import pandas as pd
import numpy as np

from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("iris.csv")
iris.sample(10)
```

```python
iris.describe()
```

No null values. that's nice

```python
print(iris["Species"].value_counts())
```

The above result shows there are only three types of species included into the dataset and each specy is equi-sampled

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder

num_cols = ["SepalLengthCm","SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]
cat_cols = ["Species"]
preprocess = ColumnTransformer(
    [
    ("scaler", StandardScaler(), num_cols),
    ("label_encoding", OrdinalEncoder(), cat_cols)
    ],
    remainder='passthrough')
t = preprocess.fit_transform(iris)
passthrough_cols = [col for col in iris.columns if col not in (num_cols + cat_cols)]

t_iris = pd.DataFrame(t, columns = num_cols + cat_cols + passthrough_cols)
t_iris
```

preprocessing the data as i am going to use svm better to scale the data and also encoding the categorical column "Species" 

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size=0.2, random_state = 42)

for train_ind, test_ind in split.split(t_iris, t_iris["Species"]):
    iris_train = t_iris.loc[train_ind]
    iris_test = t_iris.loc[test_ind]

print("below is for original set --------")
print(t_iris["Species"].value_counts() / len(t_iris))
print("below is for test set ------------")
print(iris_test["Species"].value_counts() / len(iris_test))
```

```python
corr_matrix = iris_train.corr()
print(corr_matrix)
sns.heatmap(corr_matrix,annot =True)
```

```python
from matplotlib.patches import Patch

labels = preprocess.named_transformers_["label_encoding"].categories_[0]
colors = [ plt.cm.viridis(i/2) for i in range(3)] 
legend_handles = [Patch(color = colors[i], label = labels[i]) for i in range(3)]

plt.scatter(iris_train["SepalLengthCm"], iris_train["SepalWidthCm"], c = iris_train["Species"], cmap = 'viridis'
           )
plt.title("Scatterplot based on Sepal properties")
plt.xlabel("sepal_length");
plt.ylabel("sepal_width");
plt.legend(handles = legend_handles, title="Species")
plt.show()
```

The above scatter plot is based on sepallen and sepal_width only as we can see we can separate setosa from versicolor and virginica easily on the basis of sepal_only using linear_svc but we can't separate versicolor and virgini

```python
# labels = preprocess.named_transformers_["label_encoding"].categories_[0]
# colors = [ plt.cm.viridis(i/2) for i in range(3)] 
# legend_handles = [Patch(color = colors[i], label = labels[i]) for i in range(3)]

plt.scatter(iris_train["PetalLengthCm"], iris_train["PetalWidthCm"], c = iris_train["Species"], cmap = 'viridis'
           )
plt.title("Scatterplot based on petal properties")
plt.xlabel("petal_length");
plt.ylabel("petal_width");
plt.legend(handles = legend_handles, title="Species")
plt.show()
```

it's better we can separate versicolor and virginica baed on softmargin svm. before concluding lets try some more patterns

```python
col = iris_train.columns[0:4]

fig, ax = plt.subplots(2,3, figsize=(8,6))

a = b= 0
for x in range(4):
    for y in range(x+1, 4):
        if(col[x]==col[y]):
            continue
            # print(col[x], col[y])
        ax[a,b].scatter(t_iris[col[x]], t_iris[col[y]], c = t_iris["Species"], cmap = 'viridis')
        # ax[a,b].set_title("Scatterplot based on {}, {} properties".format(col[x], col[y]))
        ax[a,b].set_xlabel(col[x]);
        ax[a,b].set_ylabel("{}".format(col[y]));
        b+=1
        if (b==3):
            a+=1
            b = 0
plt.legend(handles = legend_handles, title="Species", bbox_to_anchor=(1.02, 0.5),framealpha = 0.7)
plt.tight_layout()
plt.subplots_adjust(right = 0.8)
plt.show()
```

As from figures we can choose a figure near linearly separable i am choosing (2,3) petalwidth vs petallength

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_predict

svm_clf = SVC(kernel = "linear", decision_function_shape="ovr")
svm_clf
```

```python

x_train = iris_train.iloc[:,0:4]
y_train = iris_train["Species"]

y_train_pred = cross_val_predict(svm_clf, x_train, y_train, cv = 6)
print(y_train_pred )

from sklearn.metrics import confusion_matrix, precision_score, recall_score

conf_mat = confusion_matrix(y_train, y_train_pred)
print("precision: ", precision_score(y_train, y_train_pred, average = None ), "\nRecall : ", recall_score(y_train, y_train_pred, average = None)) 
sns.heatmap(conf_mat, annot = True)
```

```python
x_train = iris_train.loc[:,["PetalLengthCm", "PetalWidthCm"]]
y_train = iris_train["Species"]

y_train_pred = cross_val_predict(svm_clf, x_train, y_train, cv = 6)
print(y_train_pred )

from sklearn.metrics import confusion_matrix, precision_score, recall_score

conf_mat = confusion_matrix(y_train, y_train_pred)
print("precision: ", precision_score(y_train, y_train_pred, average = None ), "\nRecall : ", recall_score(y_train, y_train_pred, average = None)) 
sns.heatmap(conf_mat, annot = True)
```
